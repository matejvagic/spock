```python
import os
import shutil
```

```python
os.getcwd()
```

```python
backup_directory = "datasets/children/original_labels"
labels_directory = "datasets/children/labels"
os.mkdir(backup_directory)
```

```python
for filename in os.listdir(labels_directory):
    current = os.path.join(labels_directory, filename)
    dest = os.path.join(backup_directory, filename)
    shutil.move(current, dest)
```

```python
for filename in os.listdir(backup_directory):
    print(filename)
    if (filename[0] == "."):
        continue
    with open(os.path.join(backup_directory, filename), "r") as fr:
        lines = fr.readlines()
        with open(os.path.join(labels_directory, filename), "w") as fw:
            for line in lines:
                if (line[0] == "1"):
                    continue
                else:
                    fw.writelines(list(line))
            fw.close()
        fr.close()
```
